```python
# Keep coding genes
full_dana_cds = full_dana[full_dana["type"] == "CDS"]
full_dmoj_cds = full_dmoj[full_dmoj["type"] == "CDS"]
full_dsec_cds = full_dsec[full_dsec["type"] == "CDS"]
```

```python
# Unique FlyBase terms
def unique_terms(df):
    return df['Target'].dropna().apply(lambda x: x.split(" ")[0]).unique()
```

```python
import process_ids
headers = process_ids.all_terms()
```

```python
headers
```

```python
genes_present  = pd.DataFrame.from_records(headers)
```

```python
genes_present
```

```python
# Including isoforms as separate genes
genes_present["UniProt"].nunique()
```

```python
genes_present["Gene Name"] = genes_present.apply(lambda x: x["name"].split("-")[0], axis = 1)
```

```python
# Excluding isoforms
genes_present["Gene Name"].nunique()
```

```python
genes_present
```

```python
# Cleaning Annotation DFs

def clean_df(df):
    # Keep genes matched with % identity over 0.8
    cleaned = df[df["Identity"].astype("double") > 0.8]

    # Extract FlyBase ID
    cleaned['FB_ID'] = cleaned['Target'].dropna().apply(lambda x: x.split(" ")[0])

    return cleaned

```

```python
dana_cleaned = clean_df(full_dana_cds)
dsec_cleaned = clean_df(full_dsec_cds)
dmoj_cleaned = clean_df(full_dmoj_cds)
```

```python
def confident_genes(df):
    current = df.sort_values('Identity', ascending=False).drop_duplicates(['FB_ID'])[["FB_ID", "Identity", "start", "end"]]
    current = current.merge(genes_present, left_on = "FB_ID", right_on = "ID")
    return current
dana_genes = confident_genes(dana_cleaned)
dmoj_genes = confident_genes(dmoj_cleaned)
dsec_genes = confident_genes(dsec_cleaned)
```

```python
# Change "Uniprot" to "Gene Name" to see with accounting for isoforms
dmoj_genes.groupby("UniProt").count()
```

```python
# Shared Genes in Annotation Plot
from matplotlib_venn import venn3

set1 = set(dmoj_genes["Gene Name"])
set2 = set(dana_genes["Gene Name"])
set3 = set(dsec_genes["Gene Name"])

venn3([set1, set2, set3], ('D. Mojavensis', 'D. Ananassae', 'D. Sechellia'))

plt.show()
```

```python
def distribution(df):
    # Annotation Histograms by Percent Identity
    fig, ax = plt.subplots(figsize = (6,4))

    # Plot
        # Plot histogram
    df["Identity"].astype("double").plot(kind = "hist", density = True, alpha = 0.65, bins = 30) # change density to true, because KDE uses density
        # Plot KDE
    df["Identity"].astype("double").plot(kind = "kde")

    # Quantile lines
    ax.axvline(0.8, alpha = 1, ymax = 50, linestyle = ":")


    # X
    ax.set_xlabel("Proportion Identity")
    x_start, x_end = 0, 1
    ax.set_xlim(x_start, x_end)

    # Y
    ax.set_ylim(0, 5)
    ax.set_yticklabels([])
    ax.set_ylabel("")

    ax.text(0.81, 0.6, "80% Cutoff", size = 10, alpha =.8)

    # Overall
    ax.grid(False)
    ax.set_title("% Identity Distribution", size = 17, pad = 10)

    # Remove ticks and spines
    ax.tick_params(left = False, bottom = False)
    for ax, spine in ax.spines.items():
        spine.set_visible(False)
        
    plt.show()
```

```python
distribution(full_dana_cds)
```
